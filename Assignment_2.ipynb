```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("car_crashes")
```

```python
df
```

```python
df.head()
```

```python
sns.scatterplot(x="total",y="speeding",data=df)
```

inference:the scatter plot suggests a positive relationship between the total number of car crashes and the number of car crashes means higher total car crashes also tend to have a higher number of car crashes caused by speeding.

```python
sns.scatterplot(x="total",y="alcohol",data=df)
```

inference:the scatter plot suggests the total number of car crashes increases, the number of car crashes related to alcohol also tends to increase.

```python
sns.lineplot(x="total",y="speeding",data=df)
```

Inference:the line plot indicates as the total number of car crashes increases, the number of car crashes related to speeding tends to increase as well.

```python
sns.distplot(df["total"])
```

```python
sns.scatterplot(x="total",y="not_distracted",data=df)
```

Inference::the scatter plot suggests a negative relationship between the total number of car crashes and the percentage of time drivers were not distracted as higher total car crashes tend to have a lower percentage of time when drivers are not distracted.

```python
sns.lineplot(x="total",y="not_distracted",data=df)
```

```python
sns.scatterplot(x="total",y="no_previous",data=df)
```

Inference:the scatter plot suggests a positive relationship between the total number of car crashes and the number of car crashes with drivers who have no previous accidents

```python
sns.scatterplot(x="total",y="ins_premium",data=df)
```

 Inference:the scatter plot does not reveal a clear correlation or relationship between the total number of car crashes and insurance premiums.

```python
sns.scatterplot(x="total",y="ins_losses",data=df)
```

Inference:the scatter plot does not reveal a clear correlation or relationship between the total number of car crashes and insurance losses.

```python
sns.jointplot(x="total",y="speeding",data=df)
```

```python
sns.jointplot(x="total",y="alcohol",data=df)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='abbrev', y='alcohol', data=df, ci=None)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='abbrev', data=df)
```

Inference:The countplot showing the total car crashes by state reveals variations in the number of car crashes across different states. Some states have a notably higher count of car crashes, while others have a lower count.

```python
sns.relplot(data=df,x="speeding",y="total", hue="abbrev")
plt.figure(figsize=(8, 6))
```

Inference: The relational plot illustrates the relationship between speeding-related car crashes and total car crashes, with each point representing a state and color-coded by its abbreviation. It provides an overview of how speeding correlates with the total number of crashes in different states, allowing for easy comparison.
